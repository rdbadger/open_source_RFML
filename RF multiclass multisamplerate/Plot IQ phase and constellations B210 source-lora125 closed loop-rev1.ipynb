```python
lora125_1msps_complex = np.fromfile(path_lora125_b210+"lora_1msps_b210_new.sigmf-data", np.complex64)
lora125_5msps_complex = np.fromfile(path_lora125_b210+"lora_5msps_b210_new.sigmf-data", np.complex64)
lora125_10msps_complex = np.fromfile(path_lora125_b210+"lora_10msps_b210_new.sigmf-data", np.complex64)
lora125_25msps_complex = np.fromfile(path_lora125_b210+"lora_25msps_b210_new.sigmf-data", np.complex64)
```

```python
lora125_1msps_angle = np.angle(lora125_1msps_complex)
lora125_5msps_angle = np.angle(lora125_5msps_complex)
lora125_10msps_angle = np.angle(lora125_10msps_complex)
lora125_25msps_angle = np.angle(lora125_25msps_complex)
```

```python
path_plot_fig = "/home/david/sigMF_ML/RF/RF_class/plot_data/" # ACE
os.chdir(path_plot_fig)
```

```python
plt.rcParams.update({'font.size': 15})
fig1 = plt.figure()
pyplot.plot(lora125_1msps_angle[100:200], color ='darkblue')
plt.title('B210 1 msps LoRa 125 kHz phase', fontsize=15)
plt.xlabel('Time in IQ samples')
plt.ylabel('Phase in radians')
plt.grid()
plt.minorticks_on()
axes=plt.gca()
plt.subplots_adjust(left=0.15)
plt.subplots_adjust(bottom=0.20)
fig1.set_size_inches(6.0, 4.0)
fig1.savefig('lora125_phase_1msps_b210_20220420.png', format="png")
pyplot.show()
```

```python
plt.rcParams.update({'font.size': 15})
fig1 = plt.figure()
r = lora125_1msps_complex[200000:205000]
r = np.exp(1j*np.angle(r))
# r = r1 / np.linalg.norm(r1)
plt.plot(np.real(r), np.imag(r), '.',color ='darkblue')
plt.minorticks_on()
plt.title('B210 1 msps LoRa 125 kHz constellation')
plt.xlabel('I')
plt.ylabel('Q')
plt.grid(True)
plt.axis('equal')
plt.subplots_adjust(left=0.20)
plt.subplots_adjust(bottom=0.20)
# fig1.set_size_inches(6.0, 4.0)
fig1.savefig('lora_1msps_b210_constellation_20220420.png', format="png")
pyplot.show()
```
